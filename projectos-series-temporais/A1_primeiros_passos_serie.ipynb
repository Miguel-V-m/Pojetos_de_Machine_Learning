#  <font color='DarkRed'>**Criação de uma Série Temporal Aleatória com Distribuição Normal**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6 # corresponden a x=15,y=6 em os eixos coordenados
```

## <font color='OrangeRed'>**Série anual**

```python
# Criemao uma série de 1980 a 2020
np.random.seed(10)
dados1 = np.random.normal(0,1,41)
dados1
```

```python
type(dados1)
```

```python
serie = pd.Series(dados1)
```

```python
type(serie)
```

```python
serie.plot()
plt.show()
```

```python
dados1 = pd.DataFrame(dados1)
dados1.head(3)
```

```python
dados1.columns = ['valores']
dados1.head()
```

```python
dados1.shape
```

```python
dados1.describe()
```

```python
indice = pd.date_range('1980', periods = len(dados1), freq = 'Y')
indice
```

```python
serie1 = pd.Series(dados1['valores'].values, index = indice)
```

```python
type(serie1)
```

```python
serie1.plot()
plt.show()
```

```python
import scipy.stats as stats
```

```python
stats.probplot(serie1, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()
```

<font color='OrangeRed'>**Teste Shapiro-Wilk**

<font color='green'>**CRITÉRIOS:**

<font color='MidnightBlue'>**NÍVEL DE SIGNIFICÂNCIA DE 0,05 ou 5% (MAIS UTILIZADO)**

<font color='MidnightBlue'>**QUANDO p > 0,05 (distribuição normal).**

```python
e, p = stats.shapiro(serie1)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

## <font color='OrangeRed'>**Série mensal**

```python
np.random.seed(6)
dados2 = np.random.normal(0,1,72)
dados2
```

```python
type(dados2)
```

```python
dados2 = pd.DataFrame(dados2)
dados2
```

```python
dados2.columns = ['valores']
dados2.head()
```

```python
dados2.describe()
```

```python
dados2.shape
```

```python
import datetime     
```

```python
data = np.array('2015-01', dtype = np.datetime64())
data
```

```python
data = data + np.arange(72)
data
```

```python
data = pd.DataFrame(data)
data
```

```python
data.columns = ['data']
data.head()
```

```python
data.shape
```

```python
serie2 = pd.concat([data, dados2], axis=1)
serie2
```

```python
serie2 = pd.Series(serie2['valores'].values, index = serie2['data'])
serie2
```

```python
type(serie2)
```

```python
serie2.plot()
plt.show()
```

<font color='OrangeRed'>**Análise da normalidade**

```python
stats.probplot(serie2, dist="norm", plot=plt)
plt.title("Normal QQ plot")
plt.show()
```

<font color='OrangeRed'>**Teste Shapiro-Wilk**

<font color='green'>**CRITÉRIOS:**

<font color='MidnightBlue'>**NÍVEL DE SIGNIFICÂNCIA DE 0,05 ou 5% (MAIS UTILIZADO)**

<font color='MidnightBlue'>**QUANDO p > 0,05 (distribuição normal).**

```python
e, p = stats.shapiro(serie2)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```
