### <font color='OrangeRed'> **Extração de hashtag**
    
<font color='DarkSlateGray'>**Em seguida, como o Twitter usa Hashtags quase como um recurso de resumo (pelo menos no sentido de destacar ideias centrais). Então, vamos dar uma olhada em algumas das principais hashtags para cada uma das classes de sentimento. Faremos então "nuvens de palavras" para visualizar sua proeminência.**    
<font color='Teal'> **Hashtags são extraídas dos tweets originais e armazenadas em dataframes separados para cada classe. Isso é feito antes da limpeza do tweet para garantir que nenhuma informação seja perdida.**

```python
def hashtag_extract(tweet):
    hashtags = []
    for i in tweet:
        ht = re.findall(r"#(\w+)", i)
        hashtags.append(ht)
        
    hashtags = sum(hashtags, [])
    frequencia = nltk.FreqDist(hashtags)
    
    hashtag_df = pd.DataFrame({'hashtag': list(frequencia.keys()),'count': list(frequencia.values())})
    hashtag_df = hashtag_df.nlargest(15, columns="count")                           
    return hashtag_df 

# Extraindo os hashtags dos tweets em cada classe
pro = hashtag_extract(df2['message'][df2['sentiment'] == 'Pro'])
anti = hashtag_extract(df2['message'][df2['sentiment'] == 'Anti'])
neutral = hashtag_extract(df2['message'][df2['sentiment'] == 'Neutral'])
news = hashtag_extract(df2['message'][df2['sentiment'] == 'News'])
anti.head()
```

<font color='DarkSlateGray'>**Esta função recebe um tweet e extrai as 15 principais hashtags usando expressões regulares. Essas hashtags são armazenadas em um dataframe separado, juntamente com uma contagem da frequência com que ocorrem**

<font color='OrangeRed'> **Limpeza de tweets**
    
<font color='DarkSlateGray'>**Removemos entidades ruidosas, como URLs, pontuações, menções, números e espaço em branco extra. Os dados são ainda normalizados convertendo todas as letras em minúsculas.**

```python
def TweetCleaner(tweet):
    # Converter tudo para minúsculas
    tweet = tweet.lower()
    
    # Remover menções   
    tweet = re.sub('@[\w]*','',tweet) 
    
    # Remover url's
    tweet = re.sub(r'https?:\/\/.*\/\w*', '', tweet)
    
    # Remover hashtags
    tweet = re.sub(r'#\w*', '', tweet)    
    
    # Remover números
    tweet = re.sub(r'\d+', '', tweet) 
    
    # Remover pontuação
    tweet = re.sub(r"[,.;':@#?!\&/$]+\ *", ' ', tweet)
    
    # Remova esse diamante engraçado
    tweet = re.sub(r"U+FFFD ", ' ', tweet)
    
    # Remover espaço em branco extra
    tweet = re.sub(r'\s\s+', ' ', tweet)
    
    # Remova o espaço na frente do tweet
    tweet = tweet.lstrip(' ')
    return tweet

# limpeza de tweets na coluna de mensagens
df2['message'] = df2['message'].apply(TweetCleaner)
df2['message'] = df2['message'].apply(TweetCleaner)

df2.head()
```

<font color='DarkSlateGray'>**Esta função usa expressões regulares para remover urls, menções, hashtags, pontuação, números e qualquer espaço em branco extra de tweets depois de converter tudo para letras minúsculas.**

 <font color='OrangeRed'> **Parts of speech tagging (POS) e lematização**
    
<font color='DarkSlateGray'> **Nesta etapa, começamos determinando o comprimento de cada tweet e armazenando essas informações em uma nova coluna. Em seguida, tokenizamos os tweets antes de realizar a marcação POS em cada palavra seguida de lematização.** 
    
- <font color='DarkSlateGray'> **Na lematização, reduzimos a palavra à forma raiz do dicionário. Por exemplo, "gatos" é convertido em "gato". A lematização é feita para evitar a criação de recursos que são semanticamente semelhantes, mas sintaticamente diferentes. A lematização é preferível à lematização, pois a lematização é um método grosseiro para catalogar palavras relacionadas; ele essencialmente corta as letras do final até que a haste seja alcançada. Isso funciona muito bem na maioria dos casos, mas infelizmente o inglês tem muitas exceções onde é necessário um processo mais sofisticado.** 
    
- <font color='DarkSlateGray'> **A tokenização, também conhecida como segmentação de palavras, quebra a sequência de caracteres em um texto localizando o limite de cada palavra, ou seja, os pontos onde uma palavra termina e outra começa. Para fins de linguística computacional, as palavras assim identificadas são frequentemente chamadas de tokens.**
    
- <font color='DarkSlateGray'> **POS (part-of-speech) atribui para as palavras partes da fala, como substantivos, adjetivos, verbos**    

```python
def lemma(df2):
    df2['length'] = df2['message'].str.len()
    df2['tokenized'] = df2['message'].apply(word_tokenize)
    df2['pos_tags'] = df2['tokenized'].apply(nltk.tag.pos_tag)
    
    def get_wordnet_pos(tag):
        if tag.startswith('J'):
            return wordnet.ADJ

        elif tag.startswith('V'):
            return wordnet.VERB

        elif tag.startswith('N'):
            return wordnet.NOUN

        elif tag.startswith('R'):
            return wordnet.ADV
    
        else:
            return wordnet.NOUN
        
    wnl = WordNetLemmatizer()
    df2['pos_tags'] = df2['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
    df2['lemmatized'] = df2['pos_tags'].apply(lambda x: [wnl.lemmatize(word, tag) for word, tag in x])
    df2['lemmatized'] = [' '.join(map(str, l)) for l in df2['lemmatized']]  
    return df2

df2 = lemma(df2)
df2.head()
```

<font color='DarkSlateGray'> **Esta função modifica o dataframe original. Uma nova coluna para o comprimento (length) de cada tweet é adicionada. Os tweets são então tokenizados e cada palavra recebe uma parte da tag de fala (part of speech tag) antes de ser lematizada**

<font color='OrangeRed'> **Frequência de palavras**
    
<font color='SaddleBrown'> **Abaixo, calcularemos a frequência de palavras para cada sentimento. criaremos dataframes separados para armazenar as 25 palavras mais frequentes e suas respectivas frequências para cada classe. Uma vez extraída essa informação, construiremos WordClouds para visualizar essas palavras. As nuvens de palavras (WordClouds) são uma abordagem popular em tarefas de NPL, aqui elas nos ajudam a visualizar e obter uma melhor compreensão do que está sendo dito em cada classe. As WordClouds transmitem importância através da opacidade, portanto, quanto mais translúcida uma palavra, menos frequentemente ela aparece.**
    
<font color='SaddleBrown'> **CountVectorizer é uma ótima ferramenta fornecida pela biblioteca scikit-learn em Python. Ele é usado para transformar um determinado texto em um vetor com base na frequência (contagem) de cada palavra que ocorre em todo o texto.**   

```python
def frequency(tweet):
    
    #Vetorizador de contagem excluindo palavras irrelevantes em inglês
    cv = CountVectorizer(stop_words='english')
    words = cv.fit_transform(tweet)
    
    #Contar as palavras nos tweets e determinar a frequência de cada palavra
    sum_words = words.sum(axis=0)
    words_freq = [(word, sum_words[0, i]) for word, i in cv.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    
    #Cria um dataframe para armazenar as 25 principais palavras e suas frequências
    frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])
    frequency = frequency.head(25)
    
    return frequency

# Extrae as 25 principais palavras de cada classe
pro_frequency = frequency(df2['lemmatized'][df2['sentiment']=='Pro'])
anti_frequency = frequency(df2['lemmatized'][df2['sentiment']=='Anti'])
news_frequency = frequency(df2['lemmatized'][df2['sentiment']=='News'])
neutral_frequency = frequency(df2['lemmatized'][df2['sentiment']=='Neutral'])
```

<font color='DarkSlateGray'> **Esta função determina a frequência de cada palavra em uma coleção de tweets e armazena as 25 palavras mais frequentes em um dataframe, ordenadas da mais para a menos frequente**

```python
# Extraímos as palavras dos tweets para as classes pró e anti-mudanças climáticas
anti_words = ' '.join([text for text in anti_frequency['word']])
pro_words = ' '.join([text for text in pro_frequency['word']])
news_words = ' '.join([text for text in news_frequency['word']])
neutral_words = ' '.join([text for text in neutral_frequency['word']])

# Cria uma nuvem de palavras(wordcolud) para a classe anti-mudanças climáticas
anti_wordcloud = WordCloud(width=800,height=500,random_state=110,max_font_size=110,background_color='white',
                           colormap="Reds").generate(anti_words)

# Cria uma nuvem de palavras(wordcolud) para a classe pró mudanças climáticas
pro_wordcloud = WordCloud(width=800,height=500,random_state=73,max_font_size=110,background_color='white',
                          colormap="Greens").generate(pro_words)

# Cria uma nuvem de palavras(wordcolud) para a classe de notícias sobre mudanças climáticas
news_wordcloud = WordCloud(width=800,height=500,random_state=0,max_font_size=110,background_color='white',
                          colormap="Blues").generate(news_words)

# Cria uma nuvem de palavras(wordcolud) para a classe de mudança climática neutra
neutral_wordcloud = WordCloud(width=800,height=500,random_state=10,max_font_size=110,background_color='white',
                          colormap="Oranges").generate(neutral_words)

pro_frequency.tail()
```

### <font color='OrangeRed'> **Reconhecimento e extração de entidades nomeadas específicas**
   
<font color='DarkSlateGray'> **Extraindo as 10 principais organizações, entidades em cada classe. Essas informações são então armazenadas em dataframes separados para cada classe. Então, este é um campo inteiro de NLP - extração de entidade. Vamos usar o spacy, uma ótima biblioteca de NLP. Devemos extrair o seguinte:**  

- <font color='DarkSlateGray'> **Pessoas**  
- <font color='DarkSlateGray'> **Regiões geopolíticas**
- <font color='DarkSlateGray'> **Organizações**    
    
<font color='DarkSlateGray'> **Para esse conjunto de dados específico, estamos analisando esses fatores, pois provavelmente há alguma relação causal entre eles. É importante ressaltar que isso pode não nos dizer como esses tweeters chegariam ao espectro de apoio, mas pode nos dizer as organizações mais focadas, regiões geopolíticas e influenciadores/pessoas em advogar a favor/contra "Mudanças Climáticas Orientadas por Humanos".**    

```python
def entity_extractor(tweet):
    def get_people(tweet):  
        words = nlp(tweet)
        people = [w.text for w in words.ents if w.label_== 'PERSON']
        return people
    
    def get_org(tweet):
        words = nlp(tweet)
        org = [w.text for w in words.ents if w.label_== 'ORG']
        return org
    
    def get_gpe(tweet):
        words = nlp(tweet)
        gpe = [w.text for w in words.ents if w.label_== 'GPE']
        return gpe
    
    # Extrai as 10 melhores pessoas
    people = tweet.apply(lambda x: get_people(x))
    people = [x for sub in people for x in sub]
    people_counter = Counter(people)
    people_count = people_counter.most_common(10)
    people, people_count = map(list, zip(*people_count))
    
    # Extrai as 10 principais organizações
    org = tweet.apply(lambda x: get_org(x)) 
    org = [x for sub in org for x in sub]
    org_counter = Counter(org)
    org_count = org_counter.most_common(10)
    org, org_count = map(list, zip(*org_count))
    
    # Extrai as 10 principais entidades geopolíticas
    gpe = tweet.apply(lambda x: get_gpe(x)) 
    gpe = [x for sub in gpe for x in sub]
    gpe_counter = Counter(gpe)
    gpe_count = gpe_counter.most_common(10)
    gpe, gpe_count = map(list, zip(*gpe_count))
    
    # Cria um dataframe para armazenar as informações
    df2 = pd.DataFrame({'people' : people})
    df2['geopolitics'] = gpe
    df2['organizations'] = org
    
    return df2

# Extrae as principais entidades para cada classe
anti_info = entity_extractor(df2['lemmatized'][df2['sentiment']=='Anti'])
pro_info = entity_extractor(df2['lemmatized'][df2['sentiment']=='Pro'])
news_info = entity_extractor(df2['lemmatized'][df2['sentiment']=='News'])
neutral_info = entity_extractor(df2['lemmatized'][df2['sentiment']=='Neutral'])
```

<font color='DarkSlateGray'>**Esta função extrai as 10 principais entidades em uma coleção de tweets. A informação é então salva em um novo dataframe**
